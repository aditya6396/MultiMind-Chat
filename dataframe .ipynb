```python
import pandas as pd

# Read all CSV files into separate DataFrames
df = pd.read_csv("file.csv")
df2 = pd.read_csv("file2.csv")
df3 = pd.read_csv("file3.csv")
df4 = pd.read_csv("file4.csv")
df5 = pd.read_csv("file5.csv")
df6 = pd.read_csv("file6.csv")
df7 = pd.read_csv("file7.csv")
df8 = pd.read_csv("file8.csv")
df9= pd.read_csv("file9.csv")
df10=pd.read_csv("file10.csv")
df11=pd.read_csv("file11.csv")

# Concatenate all DataFrames into one
frames = [df, df2, df3, df4, df5, df6, df7, df8,df10,df11]
result = pd.concat(frames)

# List of repeated questions
repeated_questions = [
    "Which feature is not a part of the Shrutlekhan-Advance system?",
    "Which organization is not mentioned as using Shrutlekhan-Advance?",
    "Which languages are supported by Shrutlekhan-Advance for transcription?",
    "What is the typical rating for the 3 Phase 4 wire STATCOM developed by C-DAC?",
    "What is the primary function of the Shrutlekhan-Advance system?",
    "What role does Dr. (Ms). P. R. Lakshmi Eswari hold at C-DAC Hyderabad?",
    "Who is responsible for Quantum Technologies at C-DAC?",
    "What accuracy range does the Shrutlekhan-Advance achieve in its evaluations?",
    "What is the role of Ms. Lakshmi Panat at C-DAC Pune?",
    "Who is the Executive Director of C-DAC Mumbai?",
    "What does Ayusoft specialize in?"
]

# Remove rows with repeated questions
result_unique = result[~result['question'].isin(repeated_questions)]

# Save to a new CSV file
result_unique.to_csv("unique_questions2.csv", index=False)

print("Unique questions saved to 'unique_questions2.csv'")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score


X_train, X_test, y_train, y_test = train_test_split(df['question'], df['category'], test_size=0.2, random_state=42)


tfidf_vectorizer = TfidfVectorizer(max_features=1000)  
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)


svm_classifier = SVC(kernel='linear')
svm_classifier.fit(X_train_tfidf, y_train)


y_pred = svm_classifier.predict(X_test_tfidf)

# Evaluate model performance
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
import xgboost as xgb
from sklearn.metrics import classification_report, accuracy_score


label_encoder = LabelEncoder()
df['CategoryEncoded'] = label_encoder.fit_transform(df['category'])


X_train, X_test, y_train, y_test = train_test_split(df['question'], df['CategoryEncoded'], test_size=0.2, random_state=42)


tfidf_vectorizer = TfidfVectorizer(max_features=1000)  
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)


dtrain = xgb.DMatrix(X_train_tfidf, label=y_train)
dtest = xgb.DMatrix(X_test_tfidf, label=y_test)


params = {
    'objective': 'multi:softmax',  
    'num_class': len(label_encoder.classes_),  
    'eval_metric': 'merror',  
    'max_depth': 6,  
    'eta': 0.3  
}


num_rounds = 200 
xgb_model = xgb.train(params, dtrain, num_rounds)


y_pred = xgb_model.predict(dtest)


y_pred = y_pred.astype(int)


print("Classification Report:")
print(classification_report(y_test, y_pred, labels=range(len(label_encoder.classes_)), target_names=label_encoder.classes_))
print("Accuracy:", accuracy_score(y_test, y_pred))*100
```
